# Using Sklearn for Ridge Regression

Cannot predict solely based on asl - need to account for budget and runtime because that has to do with DTG. Will try ridge regression on these three variables 

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge

from sklearn.cross_validation import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error
```

```python
ridge_df = merged_df[['DTG', 'Budget', 'Runtime', 'ASL']].dropna()
ridge_df['Budget'] = np.log(ridge_df['Budget'])
ridge_df['ASL'] = np.log(ridge_df['ASL'])
X_ridge = ridge_df[['Budget', 'Runtime', 'ASL']]
y_ridge = np.log(ridge_df[['DTG']])
X_ridge.head()
```

```python
x_ridge_div = X_ridge.copy()
x_ridge_div['Runtime'] = np.divide(x_ridge_div['Runtime'], 100)
x_ridge_div['Budget'] = np.divide(x_ridge_div['Budget'], 10)
x_ridge_div.head()
```

```python
poly = PolynomialFeatures(2)
X_poly2 = poly.fit_transform(X_ridge)
```

```python
# split data into train and test
poly = PolynomialFeatures(3)
X_poly = poly.fit_transform(x_ridge_div)
X_train, X_test, y_train, y_test = train_test_split(X_poly, y_ridge, test_size=0.33, random_state=44)
```

```python
# First try - alpha = 0.5 to start
alpha = 8.5
model = Ridge(alpha=alpha)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.coef_
```

```python
train_error = mean_squared_error(y_train, model.predict(X_train))
test_error = mean_squared_error(y_test, model.predict(X_test))
print train_error
print test_error
```

```python
cross_sub = len(X_poly)/4
```

```python
poly = PolynomialFeatures(2)
X_poly = poly.fit_transform(x_ridge_div)
avg_MSE = []
alphas = np.linspace(0, 100, 200, endpoint=False)
for alpha in alphas:
    MSE = []
    for i in range(20):
        X_train, X_test, y_train, y_test = train_test_split(X_poly, y_ridge, test_size=0.25, random_state=i)
        model = Ridge(alpha=alpha)
        model.fit(X_train, y_train)
        test_error = mean_squared_error(y_test, model.predict(X_test))
        MSE.append(test_error)
    avg_MSE.append(np.mean(MSE))
```

```python
plt.figure(figsize=(12,8))
plt.xlabel('alpha', fontsize=14)
plt.ylabel('Cross Validation MSE', fontsize=14)
plt.title('alpha vs. Cross Validation MSE', fontsize=14)
plt.plot(alphas, avg_MSE)
```

```python
poly = PolynomialFeatures(1)
avg_MSE = []
alphas = np.linspace(-2, 8, 20, endpoint=False)
for alpha in alphas:
    MSE = []
    for i in range(20):
        X_poly = poly.fit_transform(x_ridge_div)
        X_train, X_test, y_train, y_test = train_test_split(X_poly, y_ridge, test_size=0.25, random_state=i)
        model = Ridge(alpha=alpha)
        model.fit(X_train, y_train)
        test_error = mean_squared_error(y_test, model.predict(X_test))
        MSE.append(test_error)
    avg_MSE.append(np.mean(MSE))
print avg_MSE
```

```python
plt.figure(figsize=(12,8))
plt.xlabel('alpha', fontsize=14)
plt.ylabel('Cross Validation MSE', fontsize=14)
plt.title('alpha vs. Cross Validation MSE', fontsize=14)
plt.plot(alphas, avg_MSE)
```

```python
poly = PolynomialFeatures(2)
MSE = 100000
for i in range(20):
    X_poly = poly.fit_transform(x_ridge_div)
    X_train, X_test, y_train, y_test = train_test_split(X_poly, y_ridge, test_size=0.25, random_state=i)
    model = Ridge(alpha=8.5)
    model.fit(X_train, y_train)
    train_error = mean_squared_error(y_train, model.predict(X_train))
    test_error = mean_squared_error(y_test, model.predict(X_test))
    if test_error < MSE:
        MSE = test_error
        best_model = model
        r_sq = model.score(X_test, y_test)
        coef = model.coef_
print r_sq
print coef
```

# Each Feature individually better visualization

```python
features = ['Budget', 'Runtime', 'ASL']
for feature in features:
    MSE = 1000000
    for i in range(20):
        X_train, X_test, y_train, y_test = train_test_split(features_y[feature], y_ridge['DTG'], test_size=0.25, random_state=i)
        model = LinearRegression()
        X_train = X_train.values.reshape(len(X_train),1)
        X_test = X_test.values.reshape(len(X_test),1)
        model.fit(X_train, y_train)
        train_error = mean_squared_error(y_train, model.predict(X_train))
        test_error = mean_squared_error(y_test, model.predict(X_test))
        if test_error < MSE:
            MSE = test_error
            best_model = model
            r_sq = model.score(X_test, y_test)
            coef = model.coef_
    print r_sq
    print coef
    plt.figure(features.index(feature)+1, figsize=(12,6))
    plt.plot(X_test, best_model.predict(X_test), color='red')
    plt.scatter(X_train, y_train, color='blue')
    plt.scatter(X_test, y_test, color='green')
    plt.xlabel(feature, fontsize=14)
    plt.ylabel('DTG', fontsize=14)
    plt.title(feature + ' vs. DTG', fontsize=14)
    
```
