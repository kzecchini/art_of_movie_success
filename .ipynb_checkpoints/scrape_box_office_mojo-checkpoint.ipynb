# Scraping Box Office Mojo

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import string
import re
```

```python
url = 'http://www.boxofficemojo.com/movies/alphabetical.htm?letter=A&p=.htm'
text = requests.get(url).text
```

```python
soup = BeautifulSoup(text, 'lxml')
soup = soup.find_all('table')[3]
for link in soup.find_all('a'):
    if 'schedule' not in link.attrs['href']:
        print link.attrs['href']
```

```python
letter_category_urls = []
soup2 = BeautifulSoup(text, 'lxml')
soup2 = soup2.find("div", class_="alpha-nav-holder")
for link in soup2.find_all('a'):
    letter_category_urls.append('http://www.boxofficemojo.com' + link.get('href'))
letter_category_urls
```

```python
letter_urls = [['http://www.boxofficemojo.com/movies/alphabetical.htm?letter=' + i + '&p=.htm'] for i in string.ascii_uppercase]
letter_urls        
```

```python
for letter_url in letter_urls:
    letter_category_urls = []
    soup3 = BeautifulSoup(requests.get(letter_url[0]).text)
    soup3 = soup3.find("div", class_="alpha-nav-holder")
    for link in soup3.find_all('a'):
        letter_category_urls.append('http://www.boxofficemojo.com' + link.get('href'))
    letter_url.extend(letter_category_urls)

# append movies that start with a number
letter_urls.append(['http://www.boxofficemojo.com/movies/alphabetical.htm?letter=NUM&p=.htm'])

letter_urls
```

```python
movies_urls = []
for letter_url in letter_urls:
    for letter_category in letter_url:
        soup4 = BeautifulSoup(requests.get(letter_category).text, 'lxml')
        soup4 = soup4.find_all('table')[3]
        for link in soup4.find_all('a'):
            if 'schedule' not in link.attrs['href']:
                movies_urls.append('http://www.boxofficemojo.com' + link.get('href'))
```

```python
# testing movie urls
movies_urls[-10:]
```

```python
def get_movie_title(soup):
    title_string = soup.find('title').text
    title = title_string.split('- Box Office Mojo')[0].strip()
    has_year = (title[-1] == ')' and title[-6] == '(' and title[-5:-1].isdigit())
    while has_year:
        title = title[:-6].strip()
        has_year = (title[-1] == ')' and title[-6] == '(' and title[-5:-1].isdigit())
    return title
```

```python
def get_movie_value(soup, field_name):
    '''Grab a value from boxofficemojo HTML
    
    Takes a string attribute of a movie on the page and
    returns the string in the next sibling object
    (the value for that attribute)
    or None if nothing is found.
    '''
    obj = soup.find(text=re.compile(field_name))
    if not obj: 
        return None
    # this works for most of the values
    next_obj = obj.find_next()
    if next_obj:
        return next_obj.text 
    else:
        return None
```

```python
# applying functions to grab necessary data

def get_movie_values(soup):
    title = get_movie_title(soup)
    dtg = get_movie_value(soup, 'Domestic Total Gross:')
    prod_budget = get_movie_value(soup, 'Production Budget:')
    release_date = get_movie_value(soup, 'Release Date:')
    runtime = get_movie_value(soup, 'Runtime:')
    rating = get_movie_value(soup, 'Rating:')
    widest_release = get_movie_value(soup, 'Widest\xa0Release:')
    genre = get_movie_value(soup, 'Genre:')
    director = get_movie_value(soup, 'Director:')
    title_values = [title, dtg, prod_budget, release_date, runtime, rating, widest_release, genre, director]
    return title_values
```

```python
# testing function
leb_title = get_movie_values(BeautifulSoup(requests.get(
            'http://www.boxofficemojo.com/movies/?id=aliceinwonderland10.htm').text, 'lxml'))
leb_title
```

```python
# run through and get movie values from the movie urls

movie_dict = {}

for movie_url in movies_urls:
    soup = BeautifulSoup(requests.get(movie_url).text, 'lxml')
    movie_values = get_movie_values(soup)
    movie_dict[movie_url] = movie_values
```

```python
# create dataframe from dictionary

mojo_df = pd.DataFrame(movie_dict)
mojo_df = pd.DataFrame.transpose(mojo_df)
mojo_df = mojo_df.reset_index()
mojo_df
mojo_df.columns = ['URL', 'Title', 'Domestic Total Gross', 'Budget', 
                'Date', 'Runtime', 'Rating', 'Widest Release', 
                'Genre', 'Director']
```
